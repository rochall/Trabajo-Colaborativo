# Algebra matricial 


```python
import numpy as np

# Crear matrices
A = np.array([[1, 2],
              [3, 4]])

B = np.array([[5, 6],
              [7, 8]])

print("Matriz A:")
print(A)

print("\nMatriz B:")
print(B)
```

## Suma de matrices

```python
suma = A + B
print("\nSuma:")
print(suma)
```

## Multiplicación por escalar

```python
escalar = 3 * A
print("\nMultiplicación por escalar:")
print(escalar)
```

## Multiplicación de matrices

```python
producto = A @ B
# también: np.dot(A, B)

print("\nProducto de matrices:")
print(producto)
```

## Determinante

```python
det = np.linalg.det(A)
print("\nDeterminante:")
print(det)
```
